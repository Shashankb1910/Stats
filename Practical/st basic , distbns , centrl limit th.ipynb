```python
# qq plot == meeting theoratical expectation to real time expectation

import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab

def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample , kde=True)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(log_nrml)
```

```python
plot_data(np.log(log_nrml))
```

```python
# -------------Bernouli-------------

p=0.3

sample_size=1000

bernouli_data=np.random.choice([0,1],size=sample_size,p=[1-p,p])

sns.histplot(bernouli_data,bins=2,kde=False,color="red")
```

```python
# --------Binomial ----------
trials=10
prob=0.6

bino=np.random.binomial(trials,prob,10000)
```

```python
sns.distplot(bino)
```

```python
# --------Poisson -------------

poisson_data=np.random.poisson(lam=5, size=1000) # lam = 5 interval per event

ax=sns.distplot(poisson_data , kde = False , color='red')

ax.set(xlabel='Poisson Distribution' , ylabel = 'Frequency')

plt.show()
```
